```python
#Now, construct an education DEV subset 
keywords_to_filter_by = ["education", "chemistry", "history", "math","statistic","calculus", "learn","computer science",
                         "economics", "philosophy", "political science", "physics", "geometry", "sociology", "science", "ethics",
                         "algorithm", "robotics", "programming", "coding",
                         "electrical engineering", "computer vision", "algebra", "ecology", "gender studies", "genetics", "combinatorics", "probability", 
                         "variance", "bernoulli", "binomial", "poisson", "convolution", "derivative", "integral", "for loop", 
                         "while loop", "python", "numpy", "exponents", "cosine", "recursion", "html", "css", "javascript", "quiz", "square root", "proportion", "functions",
                         "fraction","inequalities", "linear", "circle","graph", "absolute value", "equation", "theorem", "vector", "tangent", "optimization", "asymptote", "slope", "induction", "regression", 
                         "experiment", "correlation", "decimal", "matrix", "hypothesis", "fossils", "cells", "atom", "ecosystem", "scientific","heredity", "electron", "neutron","optics", "newton",
                         "scalar","greenhouse gas", "ozone", "atmosphere", "stack", "array", "debug", "parameter", "machine learning", "torque", "decomposition", "transpose",
                         "index", "factorial", "variable", "integer", "volume", "angle", "biology", "psychology", "ancient", "architecture", "implicit", "gradient", "distribution", "theory", "principle"
                           ]
```

```python
#Create empty lists of lists that we'll convert from dataframe -> CSV
def construct_education_subset(df_name):
    new_dev_list = []                 

    idCol = df_name.iloc[:, 0]

    queryCol = df_name.iloc[:, 1]
    size = 0
    for keyword in keywords_to_filter_by:
      count = 0
      idx = 0
      #This is slow
      # for index, row in df.iterrows():
      for query in queryCol:

        #Check if the given string contains keyword
          if keyword in query:
            #Add to our new dataframe
            count+=1
            id = df_name.iloc[idx, 0]
            row_entry = [ id, query ]
            #print(row_entry)
            new_dev_list.append(row_entry)
          idx +=1
      size +=count
      print(keyword, count)

    #Now, convert this list of lists to a df -> csv
    new_dev_df = pd.DataFrame( new_dev_list )
    return new_dev_df
```

```python
new_dev_df = construct_education_subset(queries_dev_withqrels)
new_train_df = construct_education_subset(queries_train_withqrels)
```

```python
#PREPROCESS: FILTER OUT 100 OR SO DUPLICATES
new_dev_df = new_dev_df.drop_duplicates()
new_dev_df.head(1000000)
```

```python
#PROCESSING DEV AND TEST
#Now, split the dev set into a SIZE 900 TEST SET and SIZE 900 train set
new_test_df = new_dev_df.sample(n=900,random_state=4)
#And drop the sampled values from the dev set
new_dev_df_sampled = new_dev_df.drop(new_test_df.index)
# new_dev_df_sampled.head(1000000)
```

```python
#PROCESSING TRAIN TO REDUCE TO 10K 
new_train_df = new_train_df.sample(n=10000,random_state=4)
new_train_df = new_train_df.drop_duplicates()

new_train_df.head(100000)
```

```python
#PREPROCESSING TEST & OTHERS: NO DUPLICATE ENTRIES
new_test_df.drop_duplicates(subset=['query_id'])
```
