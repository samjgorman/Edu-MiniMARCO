```python
import os
import pandas as pd
import sys
```

```python
#Here, we seek to construct an education dev subset from the qrels dev set
#We begin by importing the qrels file

qrel_dev = pd.read_table('./qrels.dev.tsv')
queries_dev = pd.read_table('./queries.dev.tsv')


qrel_train = pd.read_table('./qrels.train.tsv')
queries_train = pd.read_table('./queries.train.tsv')


```

```python
queries_dev.info()
```

```python
#create column names 
qrel_dev.columns = ['query_id','ignore',"passage_id", "ignore2"]
queries_dev.columns = ['query_id','query']

qrel_train.columns = ['query_id','ignore',"passage_id", "ignore2"]
queries_train.columns = ['query_id','query']
```

```python
#FOR DEV SET
#return a dataframe that contains all queries that are contained within qrel_dev 
#filter queries_Dev keeping only values contained in qrel_dev
df2 = queries_dev
df1 = qrel_dev

#mask that filters if queries query_id is in qrel query_id
m = df2.query_id.isin(df1.query_id)
#return new queries_dev where this filter is satisfied
df2 = df2[m]
df2.head(1000000)
```

```python
#FOR TRAIN SET
#return a dataframe that contains all queries that are contained within qrel_dev 
#filter queries_Dev keeping only values contained in qrel_dev
df2 = queries_train
df1 = qrel_train

#mask that filters if queries query_id is in qrel query_id
m = df2.query_id.isin(df1.query_id)
#return new queries_dev where this filter is satisfied
train_df_withqrels = df2[m]
train_df_withqrels.head(1000000)
```

```python
#now that we have our pruned queries.dev that contain strictly values that have qrels, let's save this as a new CSV
#Be sure to delete headers
#FOR DEV SET
df2.to_csv('queries_dev_withqrels.tsv' ,sep = "\t",index=False, header=False)
```

```python
#TRAIN SET
train_df_withqrels.to_csv('queries_train_withqrels.tsv' ,sep = "\t",index=False, header=False)
```
